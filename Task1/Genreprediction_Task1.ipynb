```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report,f1_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler

```

###Loading the text files

```python
train_data = pd.read_csv("train_data.txt", sep=":::", engine='python', header=None, names=['ID', 'TITLE', 'GENRE', 'DESCRIPTION'])
train_data.head()
```

```python
test_data = pd.read_csv("test_data.txt", sep=":::", engine='python', header=None, names=['ID', 'TITLE', 'DESCRIPTION'])
test_data.head()
```

```python
test_true = pd.read_csv("test_data_solution.txt", sep=":::", engine='python', header=None, names=['ID','TITLE','GENRE', 'DESCRIPTION'])
test_true.head()
```

```python
train_data.info()
```

```python
print(train_data['ID'].nunique())
print(train_data['TITLE'].nunique())
print(train_data['GENRE'].nunique())
```

```python
train_data.drop(columns=['ID','TITLE'],inplace=True)
train_data.head()
```

```python
test_data.drop(columns=['ID','TITLE'],inplace=True)
test_data.head()
```

```python
X_train=train_data['DESCRIPTION']
y_train=train_data['GENRE']
X_test=test_data['DESCRIPTION']
y_test=test_true['GENRE']
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)
```

```python
LR_model = LogisticRegression(C=1.0, max_iter=100)  # Adjust hyperparameters as needed
LR_model.fit(X_train_tfidf,y_train)
```

```python
LR_predict = LR_model.predict(X_test_tfidf)
```

```python
accuracy = accuracy_score(y_test, LR_predict)
f1 = f1_score(y_test, LR_predict, average='weighted')
print("Accuracy for Logistic Regression:", accuracy)
print("F1-score for Logistic Regression:", f1)
```
